# Introduction to Azure Machine Learning SDK for Python

```python
from azureml.core import Workspace, Experiment, Environment, Dataset
```

## Workspace

```python
ws = Workspace.from_config()
ws.get_details()
```

## Experiment

Retrieving all `Experiments` contained within the `Workspace`:

```python
Experiment.list(ws)
```

Retrieving all time series `Experiment` runs:

```python
ts_experiment = Experiment(ws, 'AutomatedML-TimeSeriesForecasting')
experiment_runs = ts_experiment.get_runs()
for run in experiment_runs:
    print(run.id)
```

## Run

All runs in the <em>AutomatedML-TimeSeriesForecasting</em> `Experiment` can alternatively be retrieved using the `Run` class:

```python
from azureml.core.run import Run

ts_runs = Run.list(ts_experiment)
for run in ts_runs:
    print(run.id)
```

## ComputeTarget

<strong> List of supported Virtual Machines (VMs) and their respective specs: </strong>

```python
from azureml.core.compute import AmlCompute

supported_VMs = AmlCompute.supported_vmsizes(workspace = ws)
supported_VMs
```

## Environment

<strong> List of all `Environments` in the current workspace:</strong>

```python
environment = Environment.list(workspace = ws)
environment
```
