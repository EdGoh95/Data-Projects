# Data Exploration

```python
import pandas as pd
# from helper_functions import parse_xml_to_csv
```

```python
# datascience_posts_df = parse_xml_to_csv("../datascience.stackexchange.com/Posts.xml", save_path = "Data Science Posts.csv")
datascience_posts_df = pd.read_csv("Data Science Posts.csv")
```

## Data Cleaning & Preprocessing

```python
datascience_posts_df['AnswerCount'] = datascience_posts_df['AnswerCount'].fillna(-1).astype(int)
id_columns = datascience_posts_df.filter(regex = 'Id').columns
datascience_posts_df[id_columns] = datascience_posts_df[id_columns].fillna(-1).astype(int)
datascience_posts_df.set_index('Id', inplace = True, drop = False)
```

### Add mesures for the length of a post

```python
datascience_posts_df['full_text'] = datascience_posts_df['Title'].str.cat(datascience_posts_df['body_text'], sep = ' ', na_rep = '')
datascience_posts_df['text_length'] = datascience_posts_df['full_text'].str.len()
```

```python
datascience_posts_df['is_question'] = datascience_posts_df['PostTypeId'] == 1
```

## Sanity Check

```python
questions_with_accepted_answers = datascience_posts_df[
    (datascience_posts_df['is_question']) & (datascience_posts_df['AcceptedAnswerId'] != -1)]
QnA = questions_with_accepted_answers.join(datascience_posts_df[['body_text']], on ='AcceptedAnswerId', how = 'left', rsuffix = '_answer')
```

```python
pd.options.display.max_colwidth = 500
QnA[['body_text', 'body_text_answer']][:5]
```

```python
datascience_posts_df = datascience_posts_df[datascience_posts_df['PostTypeId'].isin([1, 2])]

accepted_answers = (datascience_posts_df['is_question']) & (datascience_posts_df['AcceptedAnswerId'] != -1)
received_answers = (datascience_posts_df['is_question']) & (datascience_posts_df['AnswerCount'] > 0)
no_answers = (datascience_posts_df['is_question']) & (datascience_posts_df['AnswerCount'] == 0)

print("Total number of questions: {:d} ({:d} questions with no answers, {:d} with at least one answer, {:d} with an accepted answer)".format(
    sum(datascience_posts_df['is_question']), sum(no_answers), sum(received_answers), sum(accepted_answers)))
```

## Summary Statistics

```python
datascience_posts_df.info()
```

```python
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle

datascience_posts_df['Score'] = datascience_posts_df['Score'].astype(int)
high_score = datascience_posts_df['Score'] > datascience_posts_df['Score'].median()
normal_length = datascience_posts_df['text_length'] < 2000

axe = datascience_posts_df[datascience_posts_df['is_question'] & high_score & normal_length]['text_length'].hist(
    bins = 60, density = True, histtype = 'step', color = 'orange', linewidth = 2, grid = False, figsize = (16, 10))

datascience_posts_df[datascience_posts_df['is_question'] & ~high_score & normal_length]['text_length'].hist(
    bins = 60, density = True, histtype = 'step', color = 'green', linewidth = 2, grid = False)

handles = [Rectangle((0, 0), 1, 1, color = c, ec = 'k') for c in ['orange', 'green']]
labels = ['High score', 'Low score']
plt.legend(handles, labels)
axe.set_xlabel('Sentence Length (Number of Characters)')
axe.set_ylabel('Percentage of Sentences')
```
